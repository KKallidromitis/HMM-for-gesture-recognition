```python
save='test'
path='ECE5242Proj2-test' 
```

```python
classes=['wave', 'inf', 'eight', 'circle', 'beat3', 'beat4']
features=['time','Ax','Ay','Az','Wx','Wy','Wz']
```

```python
ts_data={}

for i in listdir(path):
    
    df=pd.read_csv((path+'/'+i),header=None,sep='\t', lineterminator='\n')
    display(df)
    #i=classes[np.where([x in i for x in classes])[0][0]]
    df=df.rename(columns=dict(enumerate(features)))
    df=df.drop(['time'],axis=1)
    
    if i in ts_data:
        ts_data[i]=ts_data[i].append(df,ignore_index=True)
    else:
        ts_data[i]=df
```

```python
ts_data
```

```python
def euclidean(A,B):
    return np.sqrt(np.sum((A - B)**2, axis=1))
```

```python
def knn(data,centers):
    data=data.to_numpy(float)
    centers=centers.to_numpy(float)
    new = np.zeros((len(data)))
    for p in range(len(data)):
        new[p]=np.argmin(euclidean(data[p,:],centers))
         
    return new
```

```python
ts_new = {}
for c in listdir(path):
    ts_new[c]=pd.DataFrame(knn(ts_data[c],centers),columns=['centers'])
    ts_new[c].to_csv(save+'/'+c+'.csv',index=False)
```
